# Први задатак - Статистички шифарник општина и градова

Да бисмо знали која општина припада којој статистичкој области, а која област ком региону потребно је да имамо актуелни, званични шифарник. Он постоји на сајту [Републичког завода за статистику](https://www.stat.gov.rs/media/4255/statisticki_sifarnik-1.xls), али није у формату који једноставно можемо да прочитамо и ставимо у табелу. Ова Ексел табела је прилагођена људима и штампању. Постоји неколико, срећом једноставних, проблема које треба решити да би ова табела могла да се учита у Пајтон.

Прво, у заглављу, односно два прва реда имамо спојене ћелије због чега се број колона по редовима разликује. То онда није табела. Морамо да се решимо тих недоследности, тј. да избришемо редове који нам нису потребни.

Друго, овде се у истој колони налазе назив региона, назив области, назив града и назив општине или градске општине. То је мешање категорија. Нама су потребни подаци раздвојени по колонама где је у једној регион, у другој округ, у трећој општина итд. Исто тако, нису нам потребни сви НСТЈ кодови заједно (прва колона). Довољно је да остане кôд за област. 

Треће, на крају табеле постоје фусноте које су више за текстуални фајл него за табелу. И те редове треба да обришемо.

Ваш задатак је да направите табелу где су поменуте категорије јасно раздвојене за све општине у Србији, тј. да изгледа као на слици испод и да ту табелу снимите у CSV фајл. Тај фајл ће нам бити потребан за неке даље обраде података.

# Други задатак - Општински показатељи

На Порталу отворених података постоје [подаци о просечној нето заради по општинама у Србији](https://data.gov.rs/sr/datasets/osnovni-opshtinski-pokazatelji/). Те податке је објавио Републички завод за статистику и ставио их у фајл "Општински показатељи последње стање".

Ваш задатак је да:
- преузмете одговарајући CSV фајл,
- увезете податке у Ексел или неки други програм за табеларну обраду података,
- издвојите податке за просечну нето зараду за месец (према општини пребивалишта запослених) за 2020. годину, 11. месец,
- одредите средњу нето зараду по општинама за тај месец,
- пронађете која општина има најмању просечну нето зараду,
- нацртате хистограм просечних нето зарада и
- снимите XLSX фајл са резултатима.

Иако задатак не изгледа тежак, вероватно ће вам требати мала помоћ. Обратите пажњу на следеће:
- Немојте фајл отварати у Екселу него га тамо увезите (_import_).
- Текст је табели написан ћирилицом. То није кодни распоред који Ексел подразумева. Пробајте са UTF-8.
- Ако игде у колони постоји податак који није број, Ексел ће сматрати да је тип података у колони текст. То мора да се промени како бисмо могли да извршавамо математичке операције.
